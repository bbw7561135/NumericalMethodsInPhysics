This image shows an electronic circuit for a low-pass filter. The purpose of such a filter is to cut off the high-frequency parts of the input signal (i.e. it lets the low frequency parts of the signal pass).


Let $I$ be the current that flows through the resistor $R$ and into the capacitor $C$, with a charge $Q$ on the capacitor. Then we have
$$ IR = V_{\rm in} - V_{\rm out}, \\ Q = C V_{\rm out},\\ I = \frac{dQ}{dt}.$$
Rearranging this gives us a first-order differential equation:
$$ \frac{dV_{\rm out}}{dt} = \frac{1}{RC} ( V_{\rm in} - V_{\rm out}). $$

```python
#Define the square function so that each input t can be processed through square(t)
def square(t):
    """Square wave function to floor values and use a boolean to convert values"""
    t=np.floor(2*t)#np.floor function rounds down to closest integer
    mod = t%2 #Modulo % finds the remaindder when divided by 2, used to determine if vin is even or odd
    if mod == 0: 
        vin= 1         
    else:
        vin=-1   
    return vin #return the function
```

```python
#create an array
t=np.arange(0.0,10.1,0.1)
#Create an empty array for which the iterated square values append into an empty array
vinpoints = []
for i in t:
    vinpoints.append(square(i))
plt.plot(t,vinpoints) #plot the square wave with relevant labels
plt.title('Input voltage as a function of time for a low pass filter')
plt.xlabel('t ($\mu$s)')
plt.ylabel('Vin (V)')
plt.ylim(-2,2)
```

First value of time constant RC=0.1

```python
#Solve the differential equation using 4th order Runge Kutta Method
#define a function that implements the RHS of the differential eq
def f(vout,vin):
    ''' Definition of our numerical function
    Inputs: vout: a value or array of output voltage values,
    tp: time at which to evaluate the function
    Outputs: f(vout,tp): function evaluated for this vout and tp'''
    return (vin-vout)/RC # our function.
```

```python
#define parameters
a = 0.0 # Start of interval
b = 10.1 # end of interval
N = 1000 # number of steps. 1000 gives a nice smooth line
h = (b-a)/N # size of each step, i.e. delta t, increment = h
print "Step size h is", h
vout = 0 #initial value of vout
RC=0.1
tpoints_out = np.arange(a,b,h) # array of time points
```

Taking the same approach of Taylor expanding about different points and rearranging the results to eliminate terms we can't / don't want to calculate, it's possible to come up with even more accurate (and complicated) algorithms. In general, the fourth-order Runge Kutta (i.e. the one with errors ${\rm O}(h^5)$)is considered to have the most useful combination of simplicity and accuracy, and is the most widely used method to solve differential equations numerically.

The equations are:
$$
k_1 = h f(x,t), \\
k_2 = h f(x+\frac{1}{2}k_1, t+\frac{1}{2}h), \\
k_3 = h f(x+\frac{1}{2}k_2, t+\frac{1}{2}h), \\
k_4 = h f(x+k_3, t+h), \\
x(t+h) = x(t) + \frac{1}{6}(k_1 + 2k_2 + 2k_3 + k_4).
$$
which we implement, again as a function, in the cell below.

```python
#Solve the differential equation using 4th order Runge Kutta Method
#Define the Runge Kutta function, having a loop within the function
def RungeKutta4(tpoints_out, vout):
    '''Fourth-order Runge Kutta. Inputs: array of t-values, output voltage'''
    xpoints = [] # initialise empty output array
    for t in tpoints_out:
        xpoints.append(vout) # append the current value of vout to the solution array
        k1 = h*f(vout,square(t))
        k2 = h*f(vout+0.5*k1, square(t+0.5*h))
        k3 = h*f(vout+0.5*k2, square(t+0.5*h))
        k4 = h*f(vout+k3, square(t+h))
        vout = vout + (k1 + 2*k2 + 2*k3 + k4)/6 #value of vout for this time please
    return xpoints #return the solution array
```

```python
#call the vout values from the RK iteration function
vpointsRK4 = RungeKutta4(tpoints_out, 0.0)
#plot vout against t values with relevant labels
plt.plot(tpoints_out,vpointsRK4, label='RK4 RC=0.1')
plt.legend(loc="lower right")
plt.xlabel('t ($\mu$s)')
plt.ylabel('dVout/dt (V/$\mu$s)')
plt.title('Solution of the Low Pass Filter using the 4th order Runge Kutta Method')
```

Purpose of a low pass filter is to cut off high frequencies of the input signal. 
When the time constant is varied this affects the bandwidth of frequencies at which the signal will be cut off, as the capacitor does not have time to fully charge and discharge. 
For RC=0.01 this closely follows the input signal as the signal is very close to the Vin square wave, as the capacitor has time to fully charge. 
As RC is increased the input signal is attenuated, the greater the value of RC the greater the attentuation and the lower the bandwith of frequencies will pass through (this is shown when comparing RC=0.1 and RC=1) and hence the higher frequencies are eliminated. This is shown in the graph above showing three different values of RC and how the rate of change of output voltage signal varies with time. 

In the above graph, the input signal is a square wave, hence the variation of output voltage with time produces a discrete wave. Whereas using a sine wave as the input signal (as shown below) shows a continuous variation in the output voltage signal with respect to time. You can also clearly see the phase shift between input and output voltage when using a sine wave.
