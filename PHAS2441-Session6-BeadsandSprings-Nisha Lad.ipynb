# Normal modes of beads connected by springs: solution

You are going to calculate the normal modes of vibration of a system of $N$ beads connected by $N+1$ Hookean springs to fixed end points. The beads are able to move horizontally but not vertically. This provides us with a basic but useful model of vibrations in a solid, where the beads represent atoms and the springs the interatomic bonds.

First consider a system of just two beads, labelled 1 and 2, connected by three springs with spring constants $k_1$, $k_2$, $k_3$ to two fixed end points, as in the figure above. The displacement of each bead from its equilibrium position is given by $x_1$ and $x_2$, as shown. By considering each spring individually, we can see that the forces exerted by the springs are:

** Spring 1** exerts a force $-k_1 x_1$ on bead 1.

** Spring 2** exerts a force $-k_2(x_1 - x_2)$ on bead 1 and a force $-k_2(x_2 - x_1)$ on bead 2.

** Spring 3** exerts a force $-k_3 x_2$ on bead 2.

We can then write the total forces on the beads in the form of a matrix equation:

$$
\mathbf{F} =-\mathbf{K}\cdot \mathbf{x} = -\begin{pmatrix}
    k_1 + k_2 & -k_2 \\
    -k_2 & k_2 + k_3 
  \end{pmatrix}
  \begin{pmatrix}
   x_1 \\ x_2 \end{pmatrix}
$$

This can easily be extended to a system of $N$ beads with $N+1$ springs, which gives us a [tridiagonal](http://en.wikipedia.org/wiki/Tridiagonal_matrix) matrix for the spring constants $k_i$:

$$ \mathbf{K} = 
\begin{pmatrix}
k_1 + k_2 & -k_2 & 0 &  & \ldots &  0  \\
-k_2 & k_2 + k_3 & -k_3 & & \ldots &0 \\
 0 & -k_3 & k_3 + k_4 & -k_4 & \ldots & 0 \\
\vdots &  & \ddots & & & \vdots \\
0 & 0 & \ldots &  & -k_N & k_N + k_{N+1} \\
\end{pmatrix}
$$

We also know that for a harmonic oscillator, the spring constants are related to the angular frequency by

$$ k = m \omega^2 $$

and hence when we represent this in matrix form, we have

$$ -\mathbf{K}\mathbf{x} = -m \omega^2\mathbf{x} $$

which is an eigenvalue equation of the form $\mathbf{A} \mathbf{x} = \lambda \mathbf{x}$, where the eigenvectors of the matrix $\mathbf{K}$ will give us the bead displacements for the normal modes of vibration, and the eigenvalues the angular frequency of that mode.

Below is my code for a system of 4 beads:

I have constructed the matrix K (4x4) intially for a system of masses connected with 4 Hookean springs, where the elements of K represent the corresponding forces exerted on each mass.
I have then found the eigenvalues and eigenvectors of the matrix K, where the eigenvectors of the matrix K give us the bead displacement of the normal modes of vibration and the eigenvalues correspond to -mw^2.

From the eigenvalues I can calculate each angular frequency value corresponding to each normal mode, which I have then plotted - angular frequency as a function of mode number.

```python
### STUDENT GENERATED CELL ###
#Import relevant modules to be used within the code
import numpy as np
import numpy.linalg as la #numpy linear algebra module
import matplotlib.pyplot as plt
%matplotlib inline

#Constructing the matrix K
k1=np.linspace(2,2,4) #1-d vector containing 4 elements each with value 2
k1d=np.diag(k1) #diagonalise the vector forming a 4x4 matrix of diagonal elements 2
k2=np.linspace(-1,-1,3) #1-d vector containing 3 elements each with value -1
k2d=np.diag(k2, k=1) #diagonalise the vector offset to +1 diagonal
k3d=np.diag(k2, k=-1) #diagonalise the vector offset to -1 diagonal
K= k1d + k2d + k3d #Addition of matrices
print "The Matrix K: \n", K

#Finding the eigenvalues and eigenvectors of K
#np.linalg.eigh command finds eigenvalues and eigenvectors for hermitian matrix
eigval, eigvec = la.eigh(K)
a = np.linspace(1.0,4.0,4.0)

#Calculating the angular frequency of each normal mode and plot as a function of mode number
mode=np.arange(1,5) #Create an array
#eigenvalues correspond to -mw^2
m=1.0
w=np.sqrt(eigval/m)
#output each eigenvalue with corresponding eigenvector and angular frequency value
for i in range(len(a)):
    print "Eigenvalue", eigval[i], "corresponds to eigenvector:", eigvec[i], "and w:", w[i]
#Plot the graph with relevant labels
plt.plot(mode,w, 'bx')
plt.plot(mode,w)
plt.xlabel("Mode Number")
plt.ylabel("Angular Frequency $\omega$")
plt.title("Angular Frequency as a function of Mode Number for a system of 4 beads")
```

I have then used the eigenvectors, which give the bead displacements, and have plotted them against bead number for the first three lowest normal modes shown in the cell below.

```python
#Plot the bead displacements as a function of bead number for each of the three lowest normal modes
#define an array of x values corresponding to beadnumber
beadnum=np.arange(1,5)
#plot the first three lowest normal modes giving a label to each one against bead number
#plotting the first column of the eigenvector against beadnum
disp_mode1=eigvec[:,0]
disp_mode2=eigvec[:,1]
disp_mode3=eigvec[:,2]

plt.plot(beadnum,disp_mode1, label='1st mode')
plt.plot(beadnum,disp_mode2, label='2nd mode')
plt.plot(beadnum,disp_mode3, label='3rd mode')
print "Lowest normal mode eigenvector is:", eigvec[0]
print "2nd normal mode eigenvector is:", eigvec[1]
print "3rd normal mode eigenvector is:", eigvec[2]
#Add relevant labels to the graph
plt.xlabel("Bead Number")
plt.ylabel("Bead Displacements")
plt.title("Bead Displacements as a function of bead Number for a system of 4 beads")
plt.legend() #plot the legend
#plot a fine black dotted line at the zero-y-axis
plt.axhline(c='k',ls=':',lw=0.5)
```
