# Creating a Function

```python
#Import required modules to use within code
import numpy as np
import matplotlib.pyplot as plt
#produce all matplotlib outputs in the ipython notebook itself
%matplotlib inline

def convert(r,theta):
    """Convert polar coordinates r, theta to cartesian x, y"""
    #Define relationships between cartesian and polar coordinates
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    
    #Generates the x and y values from the defined function
    return [x,y]

#test conversion
convert(1,np.pi)
```

## The Galilean Spiral

```python
#Generates an array of values between 0 and 11*pi spaced at 0.01*pi
#Smaller the spacing, the curve will be more defined
#note that the array goes up to but does not include 11*pi
theta = np.arange(0, 11*np.pi, 0.01*np.pi)

#Define the relationship used for the Galilian Spiral
length = 0.1
r = theta**2 - length
print "Theta values", theta, 
print "r values", r

#Convert polar coordinates to cartesian and print
[x,y]=convert(r,theta)
print "x,y coordinates:",[x,y]

#Plot a graph of x against y, with title, x and y labels
plt.plot(x,y)
plt.title('The Galilean Spiral')
plt.xlabel('x')
plt.ylabel('y')
```

# Variation of l in the Galilean Spiral

```python
#Generates an array of values between 0 and 2*pi spaced at 0.01*pi
#note that the array goes up to but does not include 2*pi
theta = np.arange(0, 2*np.pi, 0.01*np.pi)

#Generating a while loop starting at 0 length until it reaches 10
length = 0
while length < 10: 
    
    r = theta**2 - length # Defining the relationship used
    [x,y]=convert(r,theta) # Conversion from polar to cartesian
    plt.plot(x,y) # Plotting graph, title and axis labels
    plt.title('The Galilean Spiral with variation of length')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(x,y, label=length) #Set each label to corresponding length
    plt.legend(loc="lower right") #Legend position specified
    length = length + 2 #Increasing in increments of 2
```
