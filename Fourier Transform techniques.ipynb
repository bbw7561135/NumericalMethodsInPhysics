In this notebook we are implementing methods of using the Discrete and Fast Fourier Transform (DFT & FFT) techniques numerically. The FFT is a particularly efficient algorithm to implement the DFT.
\The Fourier transform as given in PHAS2246:
$$ g(\omega) = \frac{1}{\sqrt{2 \pi}} \int_{-\infty}^{+\infty} f(t) e^{i\omega t} dt $$
and also that its main purpose is to transform from the time domain to the frequency domain.

We can already use numerical techniques here, for example by performing the integration numerically using the trapezoidal rule. But what happens if we don't have full knowledge of $f(t)$ - for example if the data we want to transform is discrete, digital, or sampled only at certain values of $t$? This is where the discrete Fourier transform steps in.

### 1. Transforms of basic functions

Generate and plot the following functions, sampled at $N = 1000$ evenly spaced points, and calculate and plot their Fourier coefficients (use either FFT (fast fourier transform) or the DFT (discrete fourier transform) function we created)

1) A square-wave with $f(x) = +1$ for $0<x<0.5$ and $-1$ for $0.5<x<1$, over a single period.

```python
#Import relevant modules to be used within the code
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline ensures graphs are embedded within the notebook
%matplotlib inline
```

```python
#Define the square function so that each input x can be processed through square(x)
def square(x):
    """Square wave function which will convert inputs to out a discrete square function
    input:x values
    which are then converted using an if statement
    output: f(x) square wave function"""
    if x<0.5: 
        f_x= 1         
    else:
        f_x=-1   
    return f_x #return the function
```

```python
#create an array sampled at N=1000 evenly spaced points
x=np.linspace(0,1,1000)
#Create an empty array for which the iterated square values append into an empty array
f_xpoints = []
#For every element in array x, pass it through our function square and append into the empty array
for i in x:
    f_xpoints.append(square(i))

plt.plot(x,f_xpoints) #plot the square wave with relevant labels
plt.title('Square Wave function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-2,2)
```

```python
#Now use Fast Fourier Transform to calculate the coefficients and transform our square wave
c_fft = np.fft.rfft(f_xpoints) # Recall the function -transforms the array of x values
#we're using the numpy rfft routine in the numpy fft submodule where rfft refers to the real sample functions
#Plot the function with relevant labels
plt.title('Fast Fourier Transform of Square Wave function')
plt.plot(abs(c_fft), 'bo', label='Absolute') #plot the absolute coefficients
plt.xlim(0,30)
plt.xlabel('k')
plt.ylabel('coefficient $c_k$')
plt.axhline(c="k",ls=":", lw=1.0) # draw black dotted line at y = 0
plt.legend()
```

### Explanation

Above is the Fourier Transform for the square wave function, this is what I expect as to form a square wave there is one main sine wave that dominates (above sinex is the dominant wave) added to many other sine waves with different arguments that make up smaller contributions to the resulting wave.

* Download the files "piano.txt" and "trumpet.txt" from Moodle, and save them in the same folder as your notebook. These files consist of digital samples of a single note played by a piano and a trumpet. These samples were recorded at a sampling rate of 44.1 kHz.

### FFT for a piano

```python
#Plotting the notes played by a piano from data file
p = piano[::500] #array reference which extracts every 500th point instead of plotting every point

print "The data file piano has ", len(piano), " sample points"
print "I have plotted every 500th point"
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.title("Plot of Data File Piano")
plt.axhline(c="k",ls=":", lw=1.0) # draw black dotted line at y = 0
plt.plot(p)
```

* Use the FFT to calculate the discrete Fourier Transform of each sample, and plot the first 10,000 coefficients (use separate plots, or subplots, for this).

```python
#Use the Fast Fourier Transform to calculate the discrete fourier transform from piano data
p1=np.fft.rfft(piano) #calculate Fourier coefficients of first 10,000 elements
#transform original data using the number rfft routine in the numpy fft submodule
plt.plot(abs(p1)) #plot the absolute value of these coefficients

print "We've calculated", len(p1), "Fourier (real) coefficients"
plt.title('Plot of first 10,000 Fourier coefficients using FFT on Piano data')
plt.xlabel('k')
plt.ylabel('coefficient $c_k$')
plt.xlim(0,10000)
plt.grid()
```

* The frequency $f$ is related to the Fourier coefficient index $k$ by 
$$ f = \frac{\nu k}{N}, $$where $\nu$ is the sampling rate and $N$ the total number of sampled data points. What note did the instruments play?  
    * Hint: if you use `%matplotlib qt` rather than %matplotlib inline for the coefficients plot, you get an interactive plot in an external window with a zoom function that makes it easier to explore the data. You'll need to close the external window before you can re-run the notebook cells though.
    * For the musicians among you: middle C is 261 Hz. For the non-musicians - give your result in Hz. 

```python
#Calculating which frequency the piano played
#sampling rate v =44.1kHz, data file piano has N=100,000 sample points
v=44.1*10**3
N=len(piano)
k=np.argmax(p1) #np.argmax command tells you x value at which maximum occurs
print k
f=v*k/N
print "The Piano played a frequency of", f, "Hz"
```

### FFT for a trumpet

```python
#Plotting the notes played by a trumpet from data file
tr=trumpet[::500] #array reference which extracts every 500th point instead of plotting every point

print "The data file trumpet has ", len(trumpet), " sample points"
print "I have plotted every 500th point"
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.title("Plot of Data File Trumpet")
plt.axhline(c="k",ls=":", lw=1.0) # draw black dotted line at y = 0
plt.plot(tr, 'g')
```

```python
#Use the Fast Fourier Transform to calculate the fast fourier transform from trumpet data
tr1=np.fft.rfft(trumpet) #calculate Fourier coefficients of first 10,000 elements
#transform original data using the number rfft routine in the numpy fft submodule
plt.plot(abs(tr1)) #plot the absolute value of these coefficients

print "We've calculated", len(tr1), "Fourier (real) coefficients"
plt.title('Plot of first 10,000 Fourier coefficients using FFT on Trumpet data')
plt.xlabel('k')
plt.ylabel('coefficient $c_k$')
plt.xlim(0,10000)
plt.grid()
```

```python
#Calculating which frequency the trumpet played
#sampling rate v =44.1kHz, data file trumpet has N=100,000 sample points
Nt=len(trumpet)
kt=np.argmax(tr1)
f2=v*kt/Nt
print "The Trumpet played a frequency of", f2, "Hz"
```

####Explanation####

For the Fourier Transform graph for the piano data we obtain one spike, whereas for the Fourier Transform graph for the trumpet data we obtain many spikes for the fourier coefficient index, k. This may be a result of a piano key being a pure note when played, hence a single frequency obtained when we transform the data from the time domain to the frequency domain. Whereas when a trumpet note is played the pitch and amplitude may be varied depending on the person's breath, hence many frequencies of small variations are obtained as many notes are played. This is also shown in the Amplitude against time graphs, where a piano note amplitude is large when first played which then decreases in amplitude, whereas when a trumpet note is played the amplitude stays constant depending on how the person plays the note.
