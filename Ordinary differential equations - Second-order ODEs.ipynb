The image above (download/view it from Moodle if it doesn't show up) shows a pendulum of length $L$ with a bob of mass $M$. We assume that all the mass of the system is in the bob, and that the system is rigid. The pendulum experiences a torque:
$$
  \tau_p = MgL \sin(\theta),
$$ and we also know that the torque of a rigid body is  given by:
$$
  \tau = I \alpha,
$$
where $I$ is the moment of inertia and $\alpha$ the angular acceleration. These are given by:
$$
  I = ML^2 \qquad \text{and} \qquad  \alpha = \frac{d\omega}{dt} = \frac{d^2\theta}{dt^2},
$$
where $\omega$ is the angular velocity. We can then rewrite the equation for the torque in terms of the angular position of the pendulum (note that a minus sign appears because the torque and the angle are measured in opposite directions):
$$
  \tau_p = - I \frac{d^2\theta}{dt^2}.
$$
Substituting and rearranging gives us the equation of motion for the pendulum:
$$
  \frac{d^2\theta}{dt^2} = - \frac{MgL}{I} \sin(\theta) = - \frac{g}{L}
  \sin(\theta),
$$
 which tells us that the bob accelerates towards the rest point at $\theta = 0$, and that the acceleration is dependent on the sine of the angle of displacement.

Now, if we were solving this analytically, we'd be stuck - there's no exact analytical solution to this. We'd need to make an approximation in order to continue. Use the small angle approximation, and put $\sin(\theta) \approx \theta$, which (as the name suggests) is only valid when $\theta$ is small. This transforms the problem into a linear ODE, which is simple to solve analytically.
But by using our numerical methods, we can solve this system for big angles. We can separate out into two first-order equations, as before, where the two variables are now the angle $\theta$ (c.f. position $y$) and the angular velocity $\omega$ (c.f. y-velocity)

Equation 1: $$ \frac{d \theta}{d t} = \omega$$
Equation 2: $$ \frac{d \omega}{dt} = -\frac{g}{L}\sin\theta $$

```python
#Define a function which implements both differential equations and outputs an array of theta and omega(w) values
def f(r,t): 
    '''Definition of the vector function to be solved
    for a mass subject to torque in a pendulum.
    Inputs are r (a vector) and t
    outputs: theta and omega(w) - solutions of 2 first order ODEs'''
    theta = r[0] # first element of r represents theta -angle of the particle
    w = r[1] # second element of r represents the w-angular velocity of the particle
    ftheta = w # dtheta/dt is angular velocity (omega (w))
    fw = (-g/L)*np.sin(theta) # dw/dt = (-g/L)*sin(theta)
    #function will return dtheta/dt and dw/dt in a vector
    return np.array([ftheta,fw], float) # convert theta and w to an array form and ensure it's a float
```

```python
#Define parameters
g = 9.8 # gravitational acceleration in m/s^2
L = 1 #1 metre
tstart = 0.0
tend = 20.0 # solve for a period of 20 seconds.
N = 1000 # number of points for Runge-Kutta
h = (tend - tstart)/N 
print 'step is', h

tpoints = np.arange(tstart, tend, h) #array of t points
```

Use the Runge-Kutta routine to solve for a pendulum with $L =$ 1 m released at $t = 0$ from rest with an initial angle (in radians) of $\theta = 0.9 \pi$ for a duration of 20 seconds (this angle is nearly vertical - about as far from a small angle as possible!)

```python
#rewriting the fourth order RK function to accomodate for 2 variables theta and omega(w)
def RungeKutta2d(r,tpoints):
    '''Fourth-order Runge-Kutta rewritten for two simultaneous equations
    Inputs: r: 2-component array , first component is theta, second omega(w)
            tpoints: array of time values.
    Outputs: [theta points, w points]: solutions for theta(t) and w(t), numpy arrays of same length as tpoints'''
    thetapoints = [] #initialise empty arrays
    wpoints = []
    for t in tpoints:
        thetapoints.append(r[0]) #append 1st column of r into thetapoints
        wpoints.append(r[1]) #append 2nd column of r into wpoints
        #Use 4th order Runge Kutta equations
        k1 = h*f(r,t) #pass each t value through the function f, then passed through runge kutta function
        k2 = h*f(r+0.5*k1, t+0.5*h)
        k3 = h*f(r+0.5*k2, t+0.5*h)
        k4 = h*f(r+k3, t+h)
        r = r + (k1 + 2*k2 + 2*k3 + k4)/6
        #function will return an array on theta and w points in vector form
    return np.array([thetapoints, wpoints]) # convert output to numpy array with 2 rows and N columns
```

```python
# define the initial conditions
init_theta = 0.9*np.pi # initial theta position, in radians
init_w = init_theta/tend # initial w (dtheta/dt)
r = np.array([init_theta, init_w]) #makes init_theta and init_w into a vector like [x,y]
print r
```

Plot the angle $\theta$ and the angular velocity $\omega$ on an appropriately labelled plot.

```python
#recall the runge kutta function
rsol = RungeKutta2d(r,tpoints) # this will be an array of the theta and w solutions
#plot theta as a function of time; first column of vector array r
plt.plot(tpoints,rsol[0],label="$\\theta$(t) (radians)")
#plot omega (w) as a function of time; second column of vector array r
plt.plot(tpoints,rsol[1],label="$w$(t) (radians/s)")
#plot legend and relevant labels
plt.legend(loc='best')
plt.axhline(c="k",ls=":", lw=0.5) # draw black dotted line at y = 0
plt.xlabel("time (s)")
plt.ylabel('Solution')
plt.title('Fourth Order Runge Kutta Solution to Second Order ODE for a Non-Linear Pendulum')
print np.shape(rsol) #shape of vector [i,j] with i columns and j rows
```

The graph above shows the solution to a non-linear pendulum in an undamped system, using the 4th order Runge Kutta Method. The results are as expected because as the pendulum is free to oscillate and subject to no damping force the amplitude of oscillations will remain constant with increasing time. 
As the initial theta value is decreased, we obtain results based on the small angle approximation, hence the graph of omega against time (green) becomes a smoother sinusoidal wave.

# A more realistic Pendulum - including a damping term

For this we need to define a damping coefficient $c$ (start with $c=0.01$), and to take into account the mass of the bob (use 0.1 kg). The expression for the angular acceleration then becomes
$$\frac{d \omega}{dt}  = -\frac{g}{L}\sin(\theta)  - \frac{c \omega}{M L^2}$$
First we write a function to calculate for the damped motion of the pendulum, and plot these results too.

```python
#Redefine the function so that it takes into account the damping term
#Define a function which implements both differential equations and outputs an array of theta and omega(w) values
def f(r,t): 
    '''Definition of the vector function to be solved
    for a mass subject to torque in a pendulum and a damping force.
    Inputs are r (a vector) and t
    outputs: theta and omega(w) - solutions of 2 first order ODEs'''
    theta = r[0] # first element of r represents theta -angle of the particle
    w = r[1] # second element of r represents the w-angular velocity of the particle
    ftheta = w # dtheta/dt is angular velocity (omega (w))
    fw = (-g/L)*np.sin(theta)-(c*w/(m*(L**2))) # dw/dt = (-g/L)*sin(theta) -c*w/mL^2 (including the damping term)
    #function will return dtheta/dt and dw/dt as an array in a vector form
    return np.array([ftheta,fw], float) # convert theta and w to an array form and ensure it's a float
```

```python
#Define parameters
g = 9.8 # gravitational acceleration in m/s^2
L = 1 #1 metre
m=0.1 #mass in kg
c=0.01 #damping coeffcient
tstart = 0.0
tend = 20.0 # solve for a period of 20 seconds.
N = 1000 # number of points for Runge-Kutta
h = (tend - tstart)/N 
print 'step is', h

tpoints2 = np.arange(tstart, tend, h) #array of t points from parameters
```

```python
#define the initial conditions
init_theta = 0.9*np.pi # initial theta position, in radians
init_w = init_theta/tend # initial w (dtheta/dt)
r = np.array([init_theta, init_w]) #makes init_theta and init_w into a vector like [x,y]
print r
```

```python
rsol2 = RungeKutta2d(r,tpoints) # recall the function: this will be an array of the theta and w solutions
#plot theta as a function of time; theta=first column of vector array r
plt.plot(tpoints,rsol2[0],label="$\\theta$(t) (radians)")
#plot omega (w) as a function of time; w=second column of vector array r
plt.plot(tpoints,rsol2[1],label="$w$(t) (radians/s)")
#plot legend and relevant labels
plt.legend(loc='best')
plt.axhline(c="k",ls=":", lw=0.5) # draw black dotted line at y = 0
plt.xlabel("time (s)")
plt.ylabel('Solution')
plt.title('Fourth Order Runge Kutta Solution to Second Order ODE for a Non-Linear Pendulum')
print np.shape(rsol2)
```

The graph above shows the solution to a non-linear pendulum under a damping force, using the 4th order Runge Kutta Method. The results are as expected because as the system is subject to a damping force the amplitude of oscillations will decrease with increasing time.
